```python
# Matplitlib - Red Green Blue 
# OpenCV - Blue - Green - Red
fixed_img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
img_gray = cv2.imread("cat_picture.jpg", cv2.IMREAD_GRAYSCALE)
if img is None: # If file doesn't exist
    print("Image {} hasn't been loaded".format(img))
else:
    print("Image has been loaded successfully")
    plt.imshow(img_gray, cmap="gray")
```
