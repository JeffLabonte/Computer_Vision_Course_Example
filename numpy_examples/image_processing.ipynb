```python
pic = Image.open("cat_picture.jpg")
```

```python
plt.imshow(pic_red)
```

```python
plt.imshow(pic_red, cmap="gray")
```

```python
pic_green = pic_arr.copy()

# Green Channel
pic_green = pic_green[:, : , 1]
plt.imshow(pic_green, cmap="gray") # cmap === color mapping
```
