```python
# Lighter the color, the more of a color it has
# Darker the color, the lestt of a color it has

plt.imshow(pic_red, cmap="gray")
```

```python
pic_green = pic_arr.copy()

# Green Channel 0 No Green pure black - 255 - Lighter has green
pic_green = pic_green[:, : , 1]
plt.imshow(pic_green, cmap="gray") # cmap === color mapping
```

```python
pic_blue = pic_arr.copy()

# Green Channel
pic_blue = pic_blue[:, : , 2]
plt.imshow(pic_blue, cmap="gray") # cmap === color mapping
```
